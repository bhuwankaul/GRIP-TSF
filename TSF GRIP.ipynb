```python
kmeans.labels_
```

```python
# USING ELBOW METHOD TO FIND THE OPTIMAL NUMBERS OF CLUSTERS.
```

```python
arr=[]
clusters=[2,3,4,5,6,7]
for cluster in clusters:
    kmean = KMeans(n_clusters=cluster, max_iter=50)
    kmean.fit(df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']])
    arr.append(kmean.inertia_)
```

```python
plt.plot(clusters,arr)
```

```python
Kmean = KMeans(n_clusters = 3,max_iter = 50,random_state = 42)
Kmean.fit(df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']])
```

```python
Kmean.labels_
```

```python
df['cluster_id']=Kmean.labels_
```

```python
df
```

```python
#Representation using cluster_id
```
